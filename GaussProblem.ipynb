```python
x = np.linspace(0, 2 * np.pi, 100)
# Calculate the sine of x
y = np.sin(x)

# Plot the sine wave
plt.plot(x, y, label='sin(x)')
plt.title('Sine Wave Example')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```
